```python
import pandas as pd
import statistics
df=pd.read_csv('height.csv')
```

```python
index=df.shape[0] #to count number of rows
```

```python
k=int(input('Enter number of clusters: '))
```

```python
nc=df['value'].sample(n=k,random_state=1).tolist()# picking k number of random values and movung to list
c=[]
whileloop=0
while c!=nc:
    whileloop=whileloop+1
    c=nc
    g=[]
    for i in range(0,k,1):
        g.append([])
    d=[0]
    for i in range(1,k,1):
        d.append(0)
    for i in range(0,index,1):
        for j in range(0,k,1):
            d[j]=df['value'][i]-c[j]
            if d[j]<0:
                d[j]=d[j]*-1
        mini=min(d)
        for j in range(0,k,1):
            if(d[j]==mini):
                ind=j
        g[ind].append(df['value'][i])
    m=[0]
    for i in range(1,k,1):
        m.append(0)
    for i in range(0,k,1):
        m[i]=statistics.mean(g[i])
    nc=m
    print('Atempt ',whileloop,': ')
    print('K Number of Clusters: ',g,'Old Centroides: ',c,'New Centroides: ',nc)
    if c==nc:
         print('As Old Centroides are now equal to New Centroides therefore now data is accurately divide into K-Clusters.')
```
